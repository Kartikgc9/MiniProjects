**Installing and Importing Libraries and installing**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
import warnings
warnings.simplefilter("ignore")
```

**Importing dataset and checking for EDA**

**Defining variables**

```python
IDV = tdf.loc[:, ['age', 'sex', 'cp', 'trtbps', 'chol', 'fbs', 'restecg', 'thalachh', 'exng', 'oldpeak', 'slp', 'caa', 'thall', 'patient_id']]
OUT = tdf['output']
IDV.head()
```

```python
IDV_train, IDV_test, OUT_train, OUT_test = train_test_split(IDV, OUT, test_size = 0.2, random_state = 42)
```

**Defining Model**

```python
Model = AdaBoostClassifier(n_estimators=100, random_state=42)
```

```python
Model.fit(IDV_train,OUT_train) 
```

**Training model**

```python
PRED = Model.predict(IDV_train)
ACC1 = accuracy_score(PRED,OUT_train)
```

```python
print('ACCURACY   >', ACC1)
```

```python
PREDTEST = Model.predict(IDV_test)
ACCTD = accuracy_score(PREDTEST,OUT_test)
```

```python
print('ACCURACY   >', ACCTD)
```

**Importing and predicting on test data**

```python
TestD = pd.read_csv('/kaggle/input/thapar-summer-school-basic-classification/test_dataset.csv')
```

```python
TestD.head()
```

```python
Predictions = Model.predict(TestD)
print(Predictions)
```

**Making Submmission file**
